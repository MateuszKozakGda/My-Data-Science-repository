```python
## Program choinka
def tree(tree_high):
    if type(tree_high) == type(int(tree_high)):
        print("Dobrze! Narysuje choinkę o podanym przez ciebie wymiarze!")
        space_num = tree_high
        star_num = 1
        tree_fun = ""
        while  space_num != 0:
            tree_fun = space_num*" "+star_num*"*"
            space_num-=1
            star_num+=2
            print(tree_fun)
        print(" "*(tree_high-1)+"**","\n"," "*(tree_high-2)+"**")
        print("\n\nPodoba się?")
    else:
        print("Podałes zły typ danej wejsciowej!\n Dana musi byc liczba calkowitą czyli: ", 
              type(int(tree_high)), "\n a ty podles daną o typie: ", type(tree_high))
tree(10)
```

```python
from scipy.stats import kappa4
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import argus
import seaborn as sb

chi = 3
h,k = 0.5,0

x = kappa4.rvs(h,k, size=1000)
y = argus.rvs(chi, size = 1000)

plt.figure(figsize=(20,20))
plt.subplot(2,2,1)
plt.hist(x, color="r", bins="rice", align="left", density=True)
plt.title("Kappa4 distribution of n=1000")

plt.subplot(2,2,2)
plt.hist(y, color="b", bins="rice", align="right", density=True)
plt.title("Argus distribution of n=1000")
plt.show()

```

```python

plt.figure(figsize=(15,15))
plt.subplot(3,2,1)
ax1 =sb.distplot(x, bins="rice", kde=False, axlabel="x values class")
plt.title("Kappa4 distribution of x (n=1000)")

plt.subplot(3,2,2)
ax2 =sb.distplot(y, bins="rice", kde=False, axlabel="y values class", color="r")
plt.title("Argus distribution of y (n=1000)")

plt.subplot(3,2,3)
#ax1_kde = sb.distplot(x, bins="rice", hist=False, axlabel="x values class")
ax1_kde = sb.kdeplot(x, shade=True)
plt.title("Kappa4 KDE of x (n=1000)")

plt.subplot(3,2,4)
#ax2_kde = sb.distplot(y, bins="rice", hist=False, axlabel="y values class", color="r")
ax2_kde = sb.kdeplot(y, shade=True, color="r")
plt.title("Argus KDE of y (n=1000)")

plt.subplot(3,2,5)
ax_dens = sb.kdeplot(x,x, cmap="Reds", n_levels=15, shade=True, shade_lowest=True)
plt.xlabel("Kappa4")
plt.ylabel("Kappa4")
plt.title("Density KDE plot of x and x (n=1000)")

plt.subplot(3,2,6)
ax_dens = sb.kdeplot(y,y, cmap="Blues", n_levels=15, shade=True, shade_lowest=True)
plt.xlabel("Argus")
plt.ylabel("Argus")
plt.title("Density KDE plot of y and y (n=1000)")

```

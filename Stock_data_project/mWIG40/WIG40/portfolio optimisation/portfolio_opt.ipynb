```python
import pandas as pd
import sqlite3
import numpy as np
from datetime import datetime
```

```python
db_link ="/home/mateusz/Desktop/Moje_repo/My-Data-Science-repository/Stock_data_project/mWIG40/WIG40/WIG40/Wig40.db"
conn = sqlite3.connect(db_link)

df = pd.read_sql('SELECT * FROM wig40',
            conn)
```

```python
df[df["ticker"]=="11 BIT STUDIOS SPOLKA AKCYJNA"].head()
```

```python
tickers_list = df["ticker"].unique()
dates = sorted(df["date"].unique())
dates = [datetime.strptime(x, "%d-%m-%Y") for x in df["date"].unique()]
dates = sorted(dates)
dates = [datetime.strftime(x, "%d-%m-%Y") for x in dates]
test_df = pd.DataFrame(data = np.zeros((len(dates), len(tickers_list))), 
                       columns=tickers_list)                   
test_df["date"] = dates
test_df = test_df.set_index("date") 


for ticker in tickers_list:
    print(ticker)
    for date in dates:
        try:  
            if date in df["date"][df["ticker"]==ticker].values:
                test_df[ticker].loc[date] = df["close"][(df["ticker"]==ticker) & (df["date"]==date)].values
        except ValueError:
            test_df[ticker].loc[date] = df["close"][(df["ticker"]==ticker) & (df["date"]==date)].astype(object).values[0]
```

```python
test_df.to_csv("ticker_close_data.csv")
```

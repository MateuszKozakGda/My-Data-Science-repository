```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ticker_close_data.csv")
df = df.drop(columns="HUUUGE, INC.")
```

```python
df["date"] = pd.to_datetime(df["date"], format = '%d-%m-%Y')
```

```python
x = df.groupby(df["date"].map(lambda x: x.year))#.std(min_count=0)
```

```python
y = df.replace(0, np.nan)
cov_matrix = y.drop(columns=['date']).pct_change().apply(lambda x: np.log(1+x)).cov()
cov_matrix
```

```python
corr_matrix = y.drop(columns=['date']).pct_change().apply(lambda x: np.log(1+x)).corr()
corr_matrix
```

```python
ann_sd = y.drop(columns=['date']).pct_change().apply(lambda x: np.log(1+x)).std().apply(lambda x: x*np.sqrt(250))
ann_sd
```

```python
y["date"]=pd.to_datetime(df["date"], format = '%d-%m-%Y')
```

```python

ind_er = y.set_index("date").resample('Y').last().pct_change().mean()
#ind_er = y.set_index("date").pct_change().mean()
ind_er.replace("NaN", 0)
ind_er
```

```python
assets = pd.concat([ind_er, ann_sd], axis=1) # Creating a table for visualising returns and volatility of assets
assets.columns = ['Returns', 'Volatility']
```

```python
assets = assets.dropna(axis=0)
assets
```

```python
p_ret = [] # Define an empty array for portfolio returns
p_vol = [] # Define an empty array for portfolio volatility
p_weights = [] # Define an empty array for asset weights

num_assets = len(df.columns)-1
print(num_assets)
num_portfolios = 100
max_number_of_assets = 0

for portfolio in range(num_portfolios):
    if max_number_of_assets<num_assets and max_number_of_assets!=0:
        ones = np.ones(max_number_of_assets)
        zeros = np.zeros(num_assets-max_number_of_assets)
        mask = np.concatenate([ones, zeros])
        np.random.shuffle(mask)
        weights = np.random.random(num_assets)*mask
    else:
        weights = np.random.random(num_assets)
    weights = weights/np.sum(weights)
    p_weights.append(weights)

    returns = np.dot(weights, ind_er) # Returns are the product of individual expected returns of asset and its 
                                      # weights 
    p_ret.append(returns)
    var = cov_matrix.mul(weights, axis=0).mul(weights, axis=1).sum().sum()# Portfolio Variance
    sd = np.sqrt(var) # Daily standard deviation
    ann_sd = sd*np.sqrt(250) # Annual standard deviation = volatility
    p_vol.append(ann_sd)
```

```python
data = {'Returns':p_ret, 'Volatility':p_vol}

for counter, symbol in enumerate(y.set_index("date").columns.tolist()):
    #print(counter, symbol)
    data[symbol+' weight'] = [w[counter] for w in p_weights]
portfolios  = pd.DataFrame(data)
portfolios.head()
```

```python
portfolios.plot.scatter(x='Volatility', y='Returns', marker='o', s=10, alpha=0.3, grid=True, figsize=[20,20])
```

```python
min_vol_port = portfolios.iloc[portfolios['Volatility'].idxmin()]
min_vol_port
```

```python
plt.subplots(figsize=[10,10])
plt.scatter(portfolios['Volatility'], portfolios['Returns'],marker='o', s=10, alpha=0.3)
plt.scatter(min_vol_port[1], min_vol_port[0], color='r', marker='*', s=500)
```

```python
rf = 0.0289 # risk factor
optimal_risky_port = portfolios.iloc[((portfolios['Returns']-rf)/portfolios['Volatility']).idxmax()]
optimal_risky_port
```

```python
plt.subplots(figsize=(10, 10))
plt.scatter(portfolios['Volatility'], portfolios['Returns'],marker='o', s=10, alpha=0.3)
plt.scatter(min_vol_port[1], min_vol_port[0], color='r', marker='*', s=500)
plt.scatter(optimal_risky_port[1], optimal_risky_port[0], color='g', marker='*', s=500)
```
